```python
def get_product_info(answer):
    soup = BeautifulSoup(answer, 'html.parser')
    product_all = soup.find_all('li')
    name_item = []
    price_item = []
    discount_item = []
    comment_item = []
    for element in product_all:
        try:
            # Extracting product name
            product_name = element.find('lsp-product-name').text.strip()
            name_item.append( product_name )
        except AttributeError:
            name_item.append("NaN")
        try:
            # Extracting product discount
            comment_name = element.find('lsp-product-badge').text.strip()
            comment_item.append(comment_name)
        except AttributeError:
            comment_item.append("NaN")

        try:
            # Extracting product price
            product_price = element.find('lsp-product-price').text.strip()
            if len(product_price)>6 : 
                nombres = re.findall(r'\b\d+\.\d+|\d+\.-|\d+\.\b|\d+\b', product_price)
                # Convertissez les nombres en floats si nécessaire
                nombres_tmp = [nombre for nombre in nombres]
                price_item.append(nombres_tmp[1])
                discount_item.append(nombres_tmp[0])
            else :
                price_item.append(product_price)
                discount_item.append('NaN')
        except AttributeError:
            price_item.append("NaN")
            discount_item.append('NaN')
    return name_item, price_item, discount_item, comment_item

# Exemple d'utilisation :
name_item, price_item,discount_item, comment_item = get_product_info(answer)
```

```python
def convert_tab(li1 , li2, li3 , li4, item, subitem): 
    tabu = {'noms': li1, 'prix' : li2, 'ristourne' : li3 , 'comment' : li4 , 'classes' : item +" "+ subitem}
    data = pd.DataFrame(tabu)
    return data
```
